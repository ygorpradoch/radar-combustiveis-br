```python
import pandas as pd
import sys
```

lendo arquivo

```python
df = pd.read_excel("semanal-estados-desde-2013.xlsx",header=17, engine="openpyxl")
```

Vendo estrutura basica do dataframe

```python
print("DataFrame criado com ", df.shape[0], "linhas e ", df.shape[1] ," colunas")
```

tipo das colunas

```python
df.dtypes
```

Nulos reais

```python
nulos = df.isnull().sum()

nulos_pct = (nulos / len(df) * 100).round(2)

resumo_nulos = pd.DataFrame({"nulos": nulos, "pct_%": nulos_pct})
print(resumo_nulos[resumo_nulos["nulos"] > 0]
      if resumo_nulos["nulos"].sum() > 0
      else "Nenhum nulo real encontrado.")
```

Nulos disfarçados (traços, vazios e zeros suspeitos)

```python
SUSPEITOS = ["-", "--", "N/A", "n/a", "NA", "null", "NULL", " ", ""]

for col in df.select_dtypes(include="object").columns:
    contagem = df[col].isin(SUSPEITOS).sum()
    if contagem > 0:
        exemplos = df[col][df[col].isin(SUSPEITOS)].unique()[:3].tolist()
        print(f"  {col:<45} {contagem:>7,} ocorrências  ex: {exemplos}")
```

Dominio das categorias

```python
for col in df.select_dtypes(include="object").columns:
    unicos = df[col].dropna().unique()
    tamanho = len(unicos)

    print(f"\nColuna: {col} ({tamanho} valores únicos)")


    if tamanho <= 30:
        print(unicos)
    else:
        print("Primeiros",unicos[:5])
        print("Ultimos: ", unicos[-5:])
```
